```python
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

books = []

for page in range(1, 6):

    url = f"https://books.toscrape.com/catalogue/page-{page}.html"

    driver.get(url)

    products = driver.find_elements(
        By.CLASS_NAME,
        "product_pod"
    )

    for product in products:

        title = product.find_element(
            By.TAG_NAME,
            "h3"
        ).find_element(
            By.TAG_NAME,
            "a"
        ).get_attribute("title")

        price = product.find_element(
            By.CLASS_NAME,
            "price_color"
        ).text

        rating = product.find_element(
            By.CLASS_NAME,
            "star-rating"
        ).get_attribute("class").split()[-1]

        availability = product.find_element(
            By.CLASS_NAME,
            "instock"
        ).text

        books.append([
            title,
            price,
            rating,
            availability
        ])

driver.quit()

df = pd.DataFrame(
    books,
    columns=[
        "title",
        "price",
        "rating",
        "availability"
    ]
)

df.to_csv(
    "books_selenium.csv",
    index=False,
    encoding="utf-8-sig"
)

print(df.head())
print(f"\nToplam kayıt: {len(df)}")
```
