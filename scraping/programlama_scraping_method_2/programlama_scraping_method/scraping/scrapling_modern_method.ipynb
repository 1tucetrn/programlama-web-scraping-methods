```python
import pandas as pd
from scrapling import Fetcher

fetcher = Fetcher()

books = []

for page in range(1, 6):

    url = f"https://books.toscrape.com/catalogue/page-{page}.html"

    response = fetcher.get(url)

    products = response.css("article.product_pod")

    for product in products:

        title = product.css("h3 a")[0].attrib["title"]

        price = product.css(".price_color")[0].text

        rating = product.css(".star-rating")[0].attrib["class"]

        availability = product.css(".instock")[0].text.strip()

        books.append([
            title,
            price,
            rating,
            availability
        ])

df = pd.DataFrame(
    books,
    columns=[
        "title",
        "price",
        "rating",
        "availability"
    ]
)

print(df.head())

df.to_csv(
    "books_scrapling.csv",
    index=False,
    encoding="utf-8-sig"
)

print(f"Toplam kayıt: {len(df)}")
```
